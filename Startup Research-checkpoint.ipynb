```python
def col_to_snake_case_safe(df):
    df.columns = (
        df.columns
        .str.strip()  # Удаляем пробелы по краям
        .str.lower()  # Переводим в нижний регистр
        .str.replace(r'\s+', '_', regex=True)  # Заменяем пробелы на _
    )
    return df
```

### Оптимизация типов для `acquisition`

- `acquired_at` переводим в формат `datetime`.
- `id`, `acquiring_company_id`, `acquired_company_id` привести к `int32` для экономии памяти.
- `price_amount` сменить на `float64`, если будут значения с центовой точностью, иначе `int32`.

```python
def optimize_types(df, date_cols=None, id_cols=None, bool_cols=None):
    """
    Оптимизирует типы данных в DataFrame.
    
    Параметры:
        df: Исходный DataFrame
        date_cols: Список колонок с датами (преобразует в datetime64[ns])
        id_cols: Список ID-колонок (преобразует в int32/Int32)
        bool_cols: Список булевых колонок (преобразует в boolean)
    """
    # Копируем DataFrame чтобы избежать предупреждений
    df = df.copy()
    
    # Преобразование дат
    if date_cols:
        for col in date_cols:
            if col in df.columns:
                df[col] = pd.to_datetime(df[col], errors='coerce')
    
    # Преобразование ID
    if id_cols:
        for col in id_cols:
            if col in df.columns:
                # Проверяем наличие пропусков
                if df[col].isnull().any():
                    df[col] = df[col].astype('Int32')  # С поддержкой NaN
                else:
                    df[col] = df[col].astype('int32')
    
    # Преобразование булевых значений
    if bool_cols:
        for col in bool_cols:
            if col in df.columns:
                df[col] = df[col].astype('boolean')
    
    return df

# Оптимизация типов
acquisition = optimize_types(
    df=acquisition,
    date_cols=['acquired_at'],
    id_cols=['id', 'acquiring_company_id', 'acquired_company_id']
)

# Дополнительная обработка price_amount
if acquisition['price_amount'].notna().any():  # Если есть не-NaN значения
    if acquisition['price_amount'].astype(str).str.contains('\.').any():
        acquisition['price_amount'] = acquisition['price_amount'].astype('float64')
    else:
        acquisition['price_amount'] = acquisition['price_amount'].astype('int32')

# Проверяем результат
acquisition.info()
```

- Даты (`founded_at`, `closed_at`, `funded_at`) — в `datetime64`.
- Логические поля (`is_first_round`, `is_last_round`) — в `bool`.

```python
# Оптимизация типов для company_and_rounds
def optimize_company_rounds(df):
    # 1. Сначала исправим дублирование столбцов
    cols = list(df.columns)
    
    # Переименуем первый company_id (индекс 0) в main_company_id
    cols[0] = 'main_company_id' 
    
    # Переименуем второй company_id (индекс 14) в related_company_id
    cols[14] = 'related_company_id'
    
    df.columns = cols

    # 2. Преобразование ID-полей
    id_cols = ['main_company_id', 'related_company_id', 'funding_round_id']
    
    for col in id_cols:
        if col in df.columns:
            # Очистка от нечисловых символов и преобразование
            df[col] = (
                df[col].astype(str)
                .str.replace(r'[^0-9]', '', regex=True)
                .replace('', pd.NA)
            )
            df[col] = pd.to_numeric(df[col], errors='coerce').astype('Int32')
    
    # 3. Даты -> datetime64
    date_cols = ['founded_at', 'closed_at', 'funded_at']
    for col in date_cols:
        if col in df.columns:
            df[col] = pd.to_datetime(df[col], errors='coerce')
    
    # 4. Логические поля -> boolean
    bool_cols = ['is_first_round', 'is_last_round']
    for col in bool_cols:
        if col in df.columns:
            df[col] = df[col].astype('boolean')
    
    # 5. Основные числовые показатели -> float32
    float_cols = [
        'investment_rounds',
        'funding_rounds',
        'funding_total',
        'raised_amount',
        'pre_money_valuation'
    ]
    for col in float_cols:
        if col in df.columns:
            df[col] = df[col].astype('float32')
    
    return df

# Применяем преобразование
company_and_rounds = optimize_company_rounds(company_and_rounds)

# Проверяем результат
company_and_rounds.info()
```

### 2.2. Люди и их образование

```python
# 1. Подсчёт записей об образовании
edu_counts = education.groupby('person_id').size().reset_index(name='edu_records')

# 2. Объединение таблиц
merged = pd.merge(
    people[['id', 'company_id']],
    edu_counts,
    left_on='id',
    right_on='person_id',
    how='left'
)

# 3. Определение наличия образования
merged['has_education'] = merged['edu_records'].notna()

# 4. Статистика по компаниям с обработкой ошибок
company_stats = merged.groupby('company_id').agg(
    total_employees=('id', 'nunique'),
    employees_with_edu=('has_education', 'sum')
).reset_index()

# Защита от деления на ноль и бесконечных значений
company_stats['missing_rate'] = np.where(
    company_stats['total_employees'] > 0,
    1 - (company_stats['employees_with_edu'] / company_stats['total_employees']),
    0  # если нет сотрудников, считаем 0% пропусков
)

# 5. Группировка по размеру компании
bins = [0, 50, 200, 1000, np.inf]
labels = ['1-50 (малые)', '51-200 (средние)', '201-1000 (крупные)', '1001+ (очень крупные)']

company_stats['size_group'] = pd.cut(
    company_stats['total_employees'],
    bins=bins,
    labels=labels,
    right=False
)

# 6. Фильтрация некорректных значений перед анализом
valid_stats = company_stats[
    company_stats['missing_rate'].notna() & 
    np.isfinite(company_stats['missing_rate'])
]

result = valid_stats.groupby('size_group').agg(
    avg_missing_rate=('missing_rate', 'mean'),
    company_count=('company_id', 'nunique')
).reset_index()

# 7. Безопасная визуализация
plt.figure(figsize=(12, 6))

# Проверяем, есть ли данные для отображения
if not result.empty and not result['avg_missing_rate'].isna().all():
    bars = plt.bar(
        result['size_group'],
        result['avg_missing_rate'],
        color='#1f77b4',
        edgecolor='black'
    )
    
    plt.title('Доля сотрудников без данных об образовании', fontsize=14)
    plt.xlabel('Размер компании', fontsize=12)
    plt.ylabel('Доля без данных', fontsize=12)
    plt.ylim(0, 1)
    
    # Добавляем подписи только для конечных значений
    for bar in bars:
        height = bar.get_height()
        if np.isfinite(height):
            plt.text(
                bar.get_x() + bar.get_width()/2,
                height,
                f'{height:.1%}',
                ha='center',
                va='bottom'
            )
    
    plt.grid(axis='y', linestyle='--', alpha=0.5)
    plt.tight_layout()
    plt.show()
else:
    print("Нет данных для построения графика")

# 8. Вывод результатов
print("\nРезультаты анализа:")
print("| Группа компаний       | Средняя доля | Количество компаний |")
print("|-----------------------|--------------|---------------------|")
for _, row in result.iterrows():
    if np.isfinite(row['avg_missing_rate']):
        print(f"| {row['size_group']:20} | {row['avg_missing_rate']:.1%}      | {row['company_count']:19} |")

# 9. Дополнительная диагностика
print("\nДиагностика данных:")
print(f"Всего компаний: {len(company_stats)}")
print(f"Компании с некорректными данными: {len(company_stats) - len(valid_stats)}")
print(f"Общая доля пропусков: {company_stats['missing_rate'].mean():.1%}")
```

### 3.1. Объединение данных

Отберём «интересные» компании из обработанного `company_and_rounds`, соберём единую таблицу и проверим её полноту/корректность.

```python
# 1. Проверка доступных столбцов
print("Доступные столбцы в company_and_rounds:", company_and_rounds.columns.tolist())

# 2. Определение критериев отбора компаний
# Проверяем наличие необходимых столбцов
required_cols = ['status', 'main_company_id']
filter_conditions = []

if 'status' in company_and_rounds.columns:
    filter_conditions.append(company_and_rounds["status"].str.lower() == "acquired")

for col in ['investment_rounds', 'funding_rounds']:
    if col in company_and_rounds.columns:
        filter_conditions.append(company_and_rounds[col].fillna(0) > 0)

# Если есть хотя бы одно условие фильтрации
if filter_conditions:
    combined_filter = filter_conditions[0]
    for cond in filter_conditions[1:]:
        combined_filter |= cond
    candidate_companies = company_and_rounds[combined_filter].copy()
else:
    candidate_companies = company_and_rounds.copy()

# 3. Создание таблицы компаний
company_cols = [
    "main_company_id", "name", "category_code", "status",
    "founded_at", "closed_at", "domain", "network_username",
    "country_code"
]

# Оставляем только существующие столбцы
available_company_cols = [col for col in company_cols if col in company_and_rounds.columns]
companies_base = (
    candidate_companies[available_company_cols]
    .drop_duplicates(subset=["main_company_id"])
    .reset_index(drop=True)
)

print(f"Отобрано компаний: {len(companies_base)}")

# 4. Создание таблицы раундов
round_cols = [
    "main_company_id", "funding_round_id", "funded_at", 
    "funding_round_type", "raised_amount"
]

available_round_cols = [col for col in round_cols if col in company_and_rounds.columns]
rounds_sub = (
    company_and_rounds.loc[
        company_and_rounds["main_company_id"].isin(companies_base["main_company_id"]), 
        available_round_cols
    ]
    .dropna(subset=["funding_round_id"])
    .reset_index(drop=True)
)

# 5. Агрегация данных по раундам
agg_funcs = {
    'funding_round_id': 'nunique',
    'raised_amount': 'sum',
    'funded_at': ['min', 'max']
}

agg_rounds = (
    rounds_sub.groupby("main_company_id")
    .agg(agg_funcs)
    .reset_index()
)

# Упрощаем мультииндекс
agg_rounds.columns = ['main_company_id', 'n_rounds', 'total_raised', 'first_round_at', 'last_round_at']

# 6. Объединение данных
result = companies_base.merge(
    agg_rounds, 
    on="main_company_id", 
    how="left"
)

# 7. Проверка качества данных
# Проверка полноты данных
print("\nПроверка полноты данных:")
print(result.isna().mean().sort_values(ascending=False))

# Проверка согласованности
if 'funding_rounds' in company_and_rounds.columns:
    result['declared_rounds'] = (
        company_and_rounds
        .groupby('main_company_id')['funding_rounds']
        .first()
        .reindex(result['main_company_id'])
    )
    result['rounds_diff'] = result['declared_rounds'].fillna(0) - result['n_rounds'].fillna(0)
    print(f"\nНесоответствие заявленных и фактических раундов: {result['rounds_diff'].abs().mean():.1f}")

# 8. Финальная подготовка данных
result['has_funding'] = result['total_raised'] > 0
result['is_acquired'] = result['status'].str.lower() == 'acquired'

# 9. Сохранение результата
final_columns = [
    'main_company_id', 'name', 'category_code', 'status',
    'country_code', 'founded_at', 'n_rounds', 'total_raised',
    'first_round_at', 'last_round_at', 'has_funding', 'is_acquired'
]

final_result = result[[col for col in final_columns if col in result.columns]]

print(f"\nФинальный размер таблицы: {final_result.shape}")
print("\nПример данных:")
final_result.head()
```

### 3.4. Цены стартапов по категориям

Выделим категории стартапов, характеризующиеся:
- типично высокими ценами;
- наибольшим разбросом цен за стартап.

```python
# 1. Фильтрация приобретенных компаний с известной категорией и ценой
acquired_companies = final_result[
    (final_result['is_acquired']) & 
    (final_result['total_raised'] > 0) & 
    (final_result['category_code'].notna())
]

print(f"Приобретенных компаний с известной категорией: {len(acquired_companies)}")

# 2. Анализ по категориям
category_stats = acquired_companies.groupby('category_code').agg(
    company_count=('main_company_id', 'count'),
    median_price=('total_raised', 'median'),
    mean_price=('total_raised', 'mean'),
    price_std=('total_raised', 'std'),
    price_iqr=('total_raised', lambda x: x.quantile(0.75) - x.quantile(0.25)),
    price_90_10_ratio=('total_raised', lambda x: x.quantile(0.90) / x.quantile(0.10) if x.quantile(0.10) > 0 else np.nan)
).reset_index()

# Фильтрация категорий с достаточным количеством наблюдений
min_companies = 10  # минимальное количество сделок для статистической значимости
filtered_categories = category_stats[category_stats['company_count'] >= min_companies].copy()

# 3. Ранжирование категорий
# По типично высоким ценам (медиана)
top_by_median = filtered_categories.nlargest(5, 'median_price')[['category_code', 'company_count', 'median_price']]
top_by_median = top_by_median.copy()
top_by_median['median_price_m'] = top_by_median['median_price'] / 1e6

# По разбросу цен (коэффициент вариации + IQR)
filtered_categories['cv'] = filtered_categories['price_std'] / filtered_categories['mean_price']
top_by_variability = filtered_categories.nlargest(5, 'cv')[['category_code', 'company_count', 'cv', 'price_iqr']]
top_by_variability = top_by_variability.copy()
top_by_variability['price_iqr_m'] = top_by_variability['price_iqr'] / 1e6

# 4. Комбинированный рейтинг (высокие цены + высокий разброс)
filtered_categories['composite_score'] = (
    filtered_categories['median_price'] / filtered_categories['median_price'].max() +
    filtered_categories['cv'] / filtered_categories['cv'].max()
)

top_combined = filtered_categories.nlargest(5, 'composite_score')[
    ['category_code', 'company_count', 'median_price', 'cv', 'composite_score']
]
top_combined = top_combined.copy()
top_combined['median_price_m'] = top_combined['median_price'] / 1e6

# 5. Вывод результатов
print("\nТоп-5 категорий по типично высоким ценам (медиана):")
print(top_by_median[['category_code', 'company_count', 'median_price_m']].to_string(index=False))

print("\nТоп-5 категорий по разбросу цен (коэффициент вариации):")
print(top_by_variability[['category_code', 'company_count', 'cv', 'price_iqr_m']].to_string(index=False))

print("\nТоп-5 категорий для рисковых инвесторов (высокая цена + высокий разброс):")
print(top_combined[['category_code', 'company_count', 'median_price_m', 'cv', 'composite_score']].to_string(index=False))

# 6. Визуализация
# --- Топ-5 по медиане ---
plt.figure(figsize=(8,5))
top_by_median_sorted = top_by_median.sort_values("median_price_m", ascending=False)
plt.bar(top_by_median_sorted["category_code"], top_by_median_sorted["median_price_m"])
plt.title("Топ-5 категорий по типично высоким ценам (медиана)", fontsize=14)
plt.ylabel("Медианная цена сделки, млн $")
plt.xlabel("Категория")
plt.xticks(rotation=30, ha="right")
plt.show()

# --- Топ-5 по разбросу ---
plt.figure(figsize=(8,5))
top_by_variability_sorted = top_by_variability.sort_values("cv", ascending=False)
plt.bar(top_by_variability_sorted["category_code"], top_by_variability_sorted["cv"])
plt.title("Топ-5 категорий по разбросу цен (коэф. вариации)", fontsize=14)
plt.ylabel("Коэффициент вариации (CV)")
plt.xlabel("Категория")
plt.xticks(rotation=30, ha="right")
plt.show()
```

**Для анализа цен приобретённых стартапов мы выделили категории компаний, которые характеризуются:**

- **типично высокими ценами покупки** (по медиане);
- **наибольшим разбросом цен** (по коэффициенту вариации и межквартильному размаху);
- **комбинацией этих двух факторов** — категории, которые могут быть особенно интересны рисковым инвесторам.

**Почему выбран именно топ-5**
- Использование **топ-5 категорий** позволяет сохранить баланс между наглядностью и информативностью.  
- Более короткий список (например, топ-3) может упустить важные категории, где есть интересные закономерности.  
- Более длинный список (например, топ-10) снизил бы читаемость отчёта и «размыл» бы акценты.  

Таким образом, топ-5 обеспечивает достаточную статистическую значимость и при этом остаётся удобным для восприятия.

**Почему медиана**
- Величины сделок сильно варьируются, присутствуют **выбросы** (от небольших сумм до миллиардных сделок).  
- Среднее значение (mean) в таких данных не отражает «типичную» цену, так как оно сильно искажается крупными сделками.  
- **Медиана** более устойчива к выбросам и лучше показывает характерный уровень цен для категории.

**Почему коэффициент вариации (CV) и IQR**
- **Коэффициент вариации (CV = std/mean)** позволяет оценить относительный разброс цен: насколько сильно различаются сделки внутри категории по сравнению с её средним уровнем.  
- **IQR (межквартильный размах)** показывает диапазон цен для «средних» 50% компаний и даёт устойчивую оценку вариативности без влияния экстремальных выбросов.  
- Использование этих двух метрик в паре помогает учесть как относительный разброс, так и «абсолютный диапазон» сделок.

**Итог**
- Категории с высокими медианными ценами интересны инвесторам, нацеленным на **надёжные вложения**.  
- Категории с высоким разбросом привлекательны для **рисковых инвесторов**, готовых к неопределённости ради потенциально высоких доходов.  
- Комбинированный рейтинг позволяет выявить ниши, где одновременно присутствуют **высокие цены и значительный риск** — именно там может быть максимальный интерес со стороны венчурных фондов и стратегических покупателей.

```python
# 1. Подготовка данных
print("Доступные столбцы:", final_result.columns.tolist())
print("\nСтатусы компаний:")
print(final_result['status'].value_counts())

# 2. Определение столбца с раундами
if 'funding_rounds' in final_result.columns:
    rounds_col = 'funding_rounds'
elif 'n_rounds' in final_result.columns:
    rounds_col = 'n_rounds'
else:
    final_result['n_rounds'] = final_result.groupby('main_company_id')['funding_round_id'].transform('nunique')
    rounds_col = 'n_rounds'

# 3. Фильтрация и подготовка
valid_data = final_result[
    (final_result[rounds_col].notna()) & 
    (final_result[rounds_col] > 0) &
    (final_result['status'].notna())
].copy()

print(f"\nКомпаний с валидными данными: {len(valid_data)}")

# 4. Агрегация и сортировка по убыванию среднего количества раундов
status_stats = valid_data.groupby('status').agg(
    companies_count=('main_company_id', 'count'),
    mean_rounds=(rounds_col, 'mean'),
    median_rounds=(rounds_col, 'median'),
    std_rounds=(rounds_col, 'std')
).round(2).reset_index()

# Сортируем по mean_rounds по убыванию
status_stats_sorted = status_stats.sort_values('mean_rounds', ascending=False)
valid_data['status'] = pd.Categorical(valid_data['status'], 
                                    categories=status_stats_sorted['status'], 
                                    ordered=True)

print("\nСтатистика раундов по статусам (отсортировано):")
print(status_stats_sorted.to_string(index=False))

# 5. Визуализация с сортировкой
plt.figure(figsize=(14, 10))

# Боксплот для распределения (отсортированный)
plt.subplot(2, 1, 1)
sns.boxplot(data=valid_data, x='status', y=rounds_col, 
           order=status_stats_sorted['status'], showfliers=False)
plt.title('Распределение количества раундов по статусам компаний\n(отсортировано по убыванию среднего)', fontsize=14, pad=20)
plt.xlabel('Статус компании', fontsize=12)
plt.ylabel('Количество раундов финансирования', fontsize=12)
plt.xticks(rotation=45)

# Столбчатая диаграмма средних значений (отсортированная)
plt.subplot(2, 1, 2)
bars = sns.barplot(data=status_stats_sorted, x='status', y='mean_rounds', 
                  palette='viridis', order=status_stats_sorted['status'])
plt.title('Среднее количество раундов по статусам компаний\n(отсортировано по убыванию)', fontsize=14, pad=20)
plt.xlabel('Статус компании', fontsize=12)
plt.ylabel('Среднее количество раундов', fontsize=12)
plt.xticks(rotation=45)

# Добавление значений и аннотаций
for i, (_, row) in enumerate(status_stats_sorted.iterrows()):
    plt.text(i, row['mean_rounds'] + 0.1, f'{row["mean_rounds"]:.1f}', 
             ha='center', va='bottom', fontweight='bold', fontsize=11)
    plt.text(i, -0.5, f'n={row["companies_count"]}', 
             ha='center', va='top', fontsize=10, alpha=0.7)

plt.ylim(0, status_stats_sorted['mean_rounds'].max() * 1.1)
plt.tight_layout()
plt.show()
```

    - **Пороговый эффект**: После 2-го раунда резко растут шансы на IPO

**Главный вывод**: Количество раундов - надежный индикатор будущего статуса компании.

**Рекомендация для заказчика**: Фокусироваться на компаниях с 2+ завершенными раундами - они демонстрируют значительно более высокие показатели успеха.

```python
# 1. Дополнительная информация о IPO
print("Дополнительная информация о IPO:")
print(f"- Всего компаний в данных: {len(final_result):,}")
print(f"- Компаний со статусом IPO: {len(final_result[final_result['status'] == 'ipo']):,}")
print(f"- Доля компаний, достигших IPO: {len(final_result[final_result['status'] == 'ipo'])/len(final_result):.2%}")
print("\n IPO (Initial Public Offering) - первичное публичное предложение, когда компания выходит на биржу")
print("и начинает продавать свои акции широкой публике. Это считается одним из самых престижных")
print("и прибыльных видов exit для стартапов и инвесторов.")

# 2. Анализ вероятности приобретения в зависимости от количества раундов
# Создаем бинарную переменную: 1 = acquired, 0 = другие статусы
valid_data['is_acquired_binary'] = (valid_data['status'] == 'acquired').astype(int)

# Группируем по количеству раундов и считаем вероятность приобретения
acquisition_probability = valid_data.groupby(rounds_col).agg(
    total_companies=('main_company_id', 'count'),
    acquired_count=('is_acquired_binary', 'sum')
).reset_index()

acquisition_probability['acquisition_prob'] = acquisition_probability['acquired_count'] / acquisition_probability['total_companies']
acquisition_probability['cumulative_prob'] = acquisition_probability['acquisition_prob'].cumsum()

# Фильтруем для релевантного диапазона раундов (1-10, так как дальше мало данных)
acquisition_probability_filtered = acquisition_probability[acquisition_probability[rounds_col] <= 10]

# 3. Визуализация вероятности приобретения
plt.figure(figsize=(15, 5))

# График вероятности приобретения
plt.subplot(1, 2, 1)
plt.bar(acquisition_probability_filtered[rounds_col], 
        acquisition_probability_filtered['acquisition_prob'],
        color='skyblue', edgecolor='black', alpha=0.7)
plt.title('Вероятность приобретения по количеству раундов', fontsize=14, pad=20)
plt.xlabel('Количество раундов финансирования', fontsize=12)
plt.ylabel('Вероятность приобретения', fontsize=12)
plt.grid(axis='y', alpha=0.3)

# Добавляем значения на столбцы
for i, row in acquisition_probability_filtered.iterrows():
    plt.text(row[rounds_col], row['acquisition_prob'] + 0.01, 
             f'{row["acquisition_prob"]:.2f}', ha='center', fontsize=9)

# График кумулятивной вероятности
plt.subplot(1, 2, 2)
plt.plot(acquisition_probability_filtered[rounds_col], 
         acquisition_probability_filtered['cumulative_prob'],
         marker='o', linestyle='-', color='coral', linewidth=2, markersize=6)
plt.title('Кумулятивная вероятность приобретения', fontsize=14, pad=20)
plt.xlabel('Количество раундов финансирования', fontsize=12)
plt.ylabel('Кумулятивная вероятность', fontsize=12)
plt.grid(alpha=0.3)
plt.ylim(0, 1.1)

# Добавляем значения на точки
for i, row in acquisition_probability_filtered.iterrows():
    plt.text(row[rounds_col], row['cumulative_prob'] + 0.03, 
             f'{row["cumulative_prob"]:.2f}', ha='center', fontsize=9)

plt.tight_layout()
plt.show()

# 4. Ключевые метрики
print("\n Ключевые метрики вероятности приобретения:")
print(f"- Максимальная вероятность: {acquisition_probability_filtered['acquisition_prob'].max():.2%} " 
      f"({acquisition_probability_filtered.loc[acquisition_probability_filtered['acquisition_prob'].idxmax(), rounds_col]} раундов)")

print(f"- Средняя вероятность: {acquisition_probability_filtered['acquisition_prob'].mean():.2%}")
print(f"- Вероятность после 1 раунда: {acquisition_probability_filtered[acquisition_probability_filtered[rounds_col] == 1]['acquisition_prob'].values[0]:.2%}")
print(f"- Вероятность после 3 раундов: {acquisition_probability_filtered[acquisition_probability_filtered[rounds_col] == 3]['acquisition_prob'].values[0]:.2%}")

# 5. Анализ "точки перелома"
breakpoint_round = acquisition_probability_filtered.loc[acquisition_probability_filtered['acquisition_prob'].idxmax(), rounds_col]
print(f"\n Точка перелома: После {breakpoint_round} раунда вероятность приобретения максимальна")
```

- **Только 640 компаний** (1.6%) достигли IPO из 40,747
